## Sanitisation Schematic
In the sanitisation process, all unnecessary data has been removed from the event and any identifiable or possibly sensitive information has been obfuscated by hashing. Hashed data included according to RFC6920, i.e. a prefix like “ni://sha-256;”  indicates the hashing algorithm (here sha-256). The sanitised event is now public, with anyone able to see it but unable to extract sensitive information.
